## Vector Operation

```python
u = np.array([1,0])
v = np.array([0,1])
z = u + v
print(z)
#this is vector addition. 

z = u + 1
print(z)
#this is called broadcasting. the scalar '1' is added to each element in the array u 

u = [1,0]
v=[0,1]
z = u + v
print(z)
# this is basic list concatenation. see the difference in the code output
```

```python
w = np.array([1,1])
z = 2*w
print(z)
#this represents the scalar multiplication of a vector 

w = [1,1]
z = 2*w
print(z)
# this is basic list operation where multiplying the list with a number just repeats the list that many times

```

```python
# CROSS PRODUCT
a = np.array([2,3])
b= np.array([1,2])
c = np.cross(a,b)
print(c)# the result is 1k (k => z-direction)

#DOT PRODUCT
d= np.dot(a,b)
print(d)
#the vector product that you know from school is the dot product in NumPy

#ELEMENT-WISE PRODUCT
e=a*b
print(e)
```

## 2-D Arrays

* X + Y (where X,Y are 2D arrays) yield Z with element-wise addition of X and Y
* X * Y (where X,Y are 2D arrays) yield Z with element-wise product of X and Y
* k*X (X is a 2D array and k is a scalar) yields Z where every element of X is multiplied by k
* np.dot(X,Y) would perform matrix multiplication if the nxm . mxp = nxp is satisfied

```python
e = np.array([[11,12,13],[21,22,23],[31,32,33]]) #the pink [] is one dimension, the blue[] is the second dimension
print(e.shape)
print(e.size)
print(e.ndim)
print(e)
```

To access the individual elements of the array we can use [row][column] or [row,column]. Both work perfectly fine

```python
print(e[0][0])
print(e[1,2])
```

```python
f = np.array([[2,0,3],[0,1,0],[0,0,0]])
g = np.dot(e,f)
print(g)
```
