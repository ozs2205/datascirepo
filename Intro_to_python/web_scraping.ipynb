```python
import pandas as pd
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/HTMLColorCodes.html"
data = requests.get(url).text
soup = BeautifulSoup(data)
table= soup.find('table')
for row in table.find_all('tr'):
    cols = row.find_all('td')
    color_name = cols[2].string
    color_code = cols[3].string
    print("{}--->{}".format(color_name,color_code))
    
```

# Project

```python
our_url = "https://en.wikipedia.org/wiki/List_of_largest_companies_in_the_United_States_by_revenue"
page = requests.get(our_url)
# print(page.status_code) --> returns 200. means we are all good
soup = BeautifulSoup(page.text, 'html') # extract the text in html format
table = soup.find_all('table')[0] #finding the 1st table from the webpage
headers = table.find_all('th') #isolating our table headers
title_list = []
for title in headers:
    title_list.append(title.text.strip()) #extracting just the text from the headers


```

```python
#constructing our data frame using the title_list we made earlier
df=pd.DataFrame(columns=title_list)
df
```

```python
column_data = table.find_all('tr')
for row in column_data[1:]: # we added this [1:] because there was an empty first row in the table
    row_data = row.find_all('td')
    ind_row_data = [data.text.strip() for data in row_data]
    
    length = len(df)
    df.loc[length] = ind_row_data


df



   
```
