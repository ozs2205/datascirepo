# REST API


* http protocol : API requests are communicated through http responses 
* URL - uniform resource locator 
* http:// - scheme
* www.google.com - base URL 
* /images/logo.png - location on the web server

|method|description|
|------|------------|
|GET| fetches data from server|
|POST| submits data to server|
|PUT| updates data on server|
|DELETE| deletes data from server|


* import **requests** (library for http requests)
* get request has content in the url and no body
* post request has no content in the url but has a body

```python
import requests
url_get = 'http://httpbin.org/get'
payload= {"name":"Joseph", "ID":"123"}
r = requests.get(url_get, params= payload)
url_post = 'http://httpbin.org/post'
r_post = requests.post(url_post, data= payload)
print(r_post.request.body)
```
