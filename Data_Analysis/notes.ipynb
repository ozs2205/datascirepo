## Preliminary Analysis


* after loading the data, it is important that we perform a analysis of the datatype of different columns
* this is required because pandas automatically assigns data types
* but we might need particular columns to have some specific datatypes to carry out machine learning and the likes
* for this we use the **df.dtypes** to check the datatypes
* then we can use the **df.describe()** to get a summary of statistics
* this generates a statistical analysis table for numeric data columns. If we wish to get the statistical analysis for the objects also, we can pass the **df.describe(include='all'). In this we can get the details of the most frequently occuring object and its frequency
* NaN = Not a Number
* df.info() provides a concise summary of your dataframe
