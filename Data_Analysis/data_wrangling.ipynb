## Data Wrangling
it can be though of as Data Cleaning, wherein we prepare our data for further analysis. the process generally consists of the following steps (not in any order):-
* Handling Missing Values
* Data Formatting (converting one type of data to other)
* Data Normalization (centering/scaling) (different columns may have different orders of magnitude hence direct comparison may not be feasible. so we do this)
* Data Binning (making categories out of numeric data)
* Converting categorical values into numerical values

### Handling missing data

#### Data Dropping
* use the df.dropna() method
* axis = 0 drops the row containing missing values in certain columns
* axis =1 drops the entire column having missing values 


#### Data Replacement
* use the df.replace(missing_value, new_value)
* the new_value can be the mean of the column
* mean = df[column].mean()

### Data Formatting

#### Transforming Columns
* we might want to transforma column from imperial to metric units for example
* we can use the command df[columnName] = function(df[columnName])
* df.rename(columns={oldName:newName}, inplace = True) can be used to then rename the column such that it conveys correct information about the column


#### Converting Data types
* we use the df.dtypes to find the datatype
* and the df[columnName]=df[columnName].astype("int") or ...("float") ,etc to convert the datatype

### Data Normalization

* we can do Xnew = Xold/Xmax
* or min-max scaling where Xnew = Xold-Xmin/Xmax-Xmin (makes the range 0 to 1)
* or use standardization where we replace the values with their Z-score. Here the new data has a mean of 0 and std-deviation of 1


### Data Binning


to bin the data, we use 3 functions, namely :-
1. **np.linspace()**<br>
bins = np.linspace( df[columnName].min(), df[columnName].max(), numberOfBins+1)
2. a list of group names (example Low Medium High)
3. **pd.cut()**
df["column_binned"]= pd.cut(df[columnName], bins, labels=*list of group names*, include_lowest = True)
