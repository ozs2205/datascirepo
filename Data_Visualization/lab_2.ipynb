```python
df['Total'] = df[range(1980,2014)].sum(axis=1)
```

```python
eighties = range(1980,1990)
nineties = range(1990,2000)
twothous = range(2000,2014)
```

```python
df = df.sort_values(by='Total', ascending=False).head(15)
df_80 = df[eighties]
df_90 = df[nineties]
df_20 = df[twothous]
```

```python
df_80['Total'] = df_80[eighties].sum(axis=1)
df_90['Total'] = df_90[nineties].sum(axis=1)
df_20['Total'] = df_20[twothous].sum(axis=1)
```

```python
new_df = pd.concat([df_80['Total'], df_90['Total'], df_20['Total']], axis=1)
```

```python
new_df.head(5)
new_df.columns = ['80s', '90s', '2000s']
```
