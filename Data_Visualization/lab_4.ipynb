```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
years = np.arange(1980,2014)
print(years)
```

```python
residents = np.random.randint(2000,10000,size=(34,))
immigrants = np.random.randint(1000,2000, size=(34,))
```

```python
fig, ax = plt.subplots()
ax.plot(years, residents)
ax.plot(years,immigrants)
plt.show()
```
