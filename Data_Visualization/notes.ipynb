### Histograms
1. partitions numeric data into bins
2. assigns each datapoint to a bin
3. plots the number of data points in each bin for all the bins
4. **frequency distribution plot**
5. we use the count, bin_edges = np.histogram(data, bins=10(*by default it is 10*)) method to get a list of bin v/s frequency 
6. and then we run the df.plot(kind='hist', xticks = bin_edges) --> this ensures that the bins are properly depicted

### Tree Maps
1. using the plotly library we can plot treemaps
2. these are useful in showcasing hierarchial data
3. syntax :-<br>
data = {category =['category-1', 'category-1', 'category-2', 'category-3', 'category-3','category-3'],<br>
